```python
# prerequisite package imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

In this workspace, you will work with the fuel economy dataset from the previous lesson on bivariate plots.

```python
fuel_econ = pd.read_csv('fuel-econ.csv')
fuel_econ.head()
```

**Task 1**: Plot the city ('city') vs. highway ('highway') fuel efficiencies (both in mpg) for each vehicle class ('VClass'). Don't forget that vehicle class is an ordinal variable with levels {Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, Large Cars}.

```python
vehicle_classes = ['Minicompact Cars', 'Subcompact Cars',' Compact Cars',' Midsize Cars',' Large Cars']
plt.figure(figsize=(8,6))
for vclass in vehicle_classes:
    class_data = fuel_econ[fuel_econ['VClass'] == vclass]
    plt.scatter(class_data['city'], class_data['highway'], label=vclass, alpha=0.5)

plt.xlabel('City Fuel Efficiency (mpg)')
plt.ylabel('Highway Fuel Efficiency (mpg)')
plt.title('City vs. Highway Fuel Efficiencies for Each Vehicle Class')
plt.legend(title='Vehicle Classes',loc = 'upper right')
plt.grid(True)
plt.show()
```

**Task 2**: Plot the relationship between engine size ('displ', in liters), vehicle class, and fuel type ('fuelType'). For the lattermost feature, focus only on Premium Gasoline and Regular Gasoline cars. What kind of relationships can you spot in this plot?

```python
filtered_fuel = fuel_econ[fuel_econ['fuelType'].isin(['Premium Gasoline', 'Regular Gasoline'])]
plt.figure(figsize=(10,6))
sb.stripplot(x='displ', y='VClass', hue='fuelType', data=filtered_fuel, dodge=True,alpha=0.5)

plt.xlabel('Engine Size (liters)')
plt.ylabel('Vehicle Class')
plt.title('The Relationship between Engine Size,Vehicle Class and Fuel Type')

plt.grid(True)
plt.legend(title='Fuel Type', bbox_to_anchor=(1, 0.5))
plt.show()
```
